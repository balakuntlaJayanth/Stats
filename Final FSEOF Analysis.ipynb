```python
specific_model=copy.deepcopy(mutant1_models['C730_20-90'])
```

```python
target=specific_model.reactions.EX_MonensinA_e0
```

```python
max_enforced_flux=0.9 
#0.9 is the maximum Flux
number_of_results=10
#How many bins current flux and 10 intermediate points
```

```python
reference=cobra.flux_analysis.pfba(specific_model)
```

```python
reference
```

```python
ndecimals=5
initial_fluxes = reference.fluxes
initial_flux = round(initial_fluxes[target.id], ndecimals)
```

```python
target.id
```

```python
initial_flux
```

```python
best_model=copy.deepcopy(specific_model)
best_model.objective=target.id
```

```python
max_theoretical_flux = round(best_model.slim_optimize(), ndecimals)
```

```python
max_theoretical_flux
```

```python
max_flux = max_theoretical_flux * max_enforced_flux
```

```python
max_flux
#we are going to simulate up to 1.27
```

```python
levels = [initial_flux + (i + 1) * (max_flux - initial_flux) / number_of_results for i in range(number_of_results)]
```

```python
levels
```

```python
results = {reaction.id: [] for reaction in specific_model.reactions}
```

```python
results
```

```python
target
```

```python
for level in levels:
    print(level)
    target.upper_bound = level
    target.lower_bound = level
    try:
        solution = cobra.flux_analysis.pfba(specific_model)
    except:
        continue
    for reaction_id, flux in solution.fluxes.iteritems():
        results[reaction_id].append(round(flux, ndecimals))
```

```python
fseof_reactions = []
```

```python
for reaction_id, fluxes in results.items():
    if min(fluxes) * max(fluxes) >= 0:
        fseof_reactions.append(specific_model.reactions.get_by_id(reaction_id))
```

```python
results = {rea.id: results[rea.id] for rea in fseof_reactions}
```

```python
results
```

```python
df=pd.DataFrame(index=results.keys(), columns=['slope','range'])
for i in df.index:
    lr=sm.OLS(results[i],results['EX_MonensinA_e0']).fit()
    df.loc[i,'slope']=lr.params[0]
    df.loc[i,'range']=abs(max(results[i])-min(results[i]))
```
